## Suppose if you have more frequent categories, we just replace NAN with a new category

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('loan.csv', usecols=['BsmtQual','FireplaceQu','GarageType','SalePrice'])
df
```

```python
# Before adding Missing category, there are 4 categories
df['BsmtQual'].value_counts().plot.bar(color='firebrick')
```

```python
def impute_nan(df,variable):
    df[variable+'_miss_check'] = np.where(df[variable].isnull(),"Missing",df[variable])
```

```python
for feature in ['BsmtQual','FireplaceQu','GarageType']:
    impute_nan(df, feature)
```

```python
df.head(30)
```

```python
df.drop(['BsmtQual','FireplaceQu','GarageType'],axis=1)
```

```python
df.head(30)
```

```python
# After adding Missing category, 5th category "Missing is created"
df['BsmtQual_miss_check'].value_counts().plot.bar(color='darkmagenta')
```

#### All the NaN values are replacd with Missing keyword and  a new category called "Missing" is added.
