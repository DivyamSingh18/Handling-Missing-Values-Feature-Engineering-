```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('loan.csv', usecols=['BsmtQual','GarageType','SalePrice'])
df
```

```python
#Creating New Feature
df['BsmtQual_missval_check'] = np.where(df['BsmtQual'].isnull(),1,0)   
```

```python
df.head(40)
```

```python
#For frequent category
frequent = df['BsmtQual'].mode()[0]
frequent
```

```python
# replacing NaN values in BsmtQual with 'TA'
df['BsmtQual'].fillna(frequent, inplace=True)
df.head(20)
```

## Above commands can be comprised into a function as given below:

```python
def impute_nan(df,var):  
    
    #Creating New Feature
    df[var+'_missval_check'] = np.where(df['BsmtQual'].isnull(),1,0)   
   
    #For frequent category 
    frequent = df[var].mode()[0]                      
    
    #Replacing NaN values in BsmtQual with 'TA'
    df['BsmtQual'].fillna(frequent, inplace=True)
    
```

## Advantages
1.Easy to Implement <br> 
2.Captures the importance of missing value.

## Disadvantages
1.Creating Additional Features Which leads to Curse of Dimensionality.
