## Capturing NAN Values with a new Feature
The new feature is created to keep the record that which values were added manually

It works well if the data are not missing completely at random

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv',usecols=['Age','Fare','Survived'])
df
```

```python
df['Age_NAN']=np.where(df['Age'].isnull(),1,0)
df.head(10)
```

```python
df.Age.median()
df['Age'].fillna(df.Age.median(),inplace=True)
df.head(10)
```

## Advantages
1.Easy to Implement<br>
2.Captures the importance of missing value.

## Disadvantages
1.Creating Additional Features Which leads to Curse of Dimensionality.
